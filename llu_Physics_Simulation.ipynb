## **Q.1 (a) [10pts] Make a simulation of the a bouncing ball with the following characterstics.**
\begin{align}
m &=1kg, p_0=(0, 1.5), v_0=(0.3, -0.1), \gamma=0.8,
\end{align}
where $p_0$ and $v_0$ are the initial position and velocity, respectively. $\gamma$ is the coefficient of resititution of the ball. Use $\Delta t=0.001$ sec and simulate the bouncing ball during 3 sec. (Note: No frictional loss)

```python
def simulate_ball(mass, coeff_rest, init_pos, init_vel, delta_t, N=3000):
  g = -9.8
  p_hist = np.zeros((N, 2))
  v_hist = np.zeros((N, 2))
  t_hist = np.zeros(N)
  p_hist[0, :] = init_pos
  v_hist[0, :] = init_vel
  for i in range(1, N):
      p_hist[i, :] = p_hist[i - 1, :] + v_hist[i - 1, :] * delta_t
      v_hist[i, :] = v_hist[i - 1, :] + np.array([0, g]) * delta_t

      ## handle bouncing
      if p_hist[i, 1] <= 0:
          p_hist[i, 1] = 0
          v_hist[i, 1] = -coeff_rest * v_hist[i, 1]

      t_hist[i] = i*delta_t;
  return p_hist, v_hist, t_hist
```

```python
# Your code here: set up proper parameters and initial state to call the 'simulate_ball' function
[p_hist, v_hist, t_hist] = simulate_ball(mass=1, coeff_rest=0.8, init_pos=np.array([0, 1.5]), init_vel=np.array([0.3, -0.1]), delta_t=0.001)
```

```python
# Bouncing ball visualization: No need to change
def plot_bouncingball(p_history, t_history, num_frames= 100):
  fig= plt.figure(figsize=(10,10))
  ax = plt.subplot(1,1,1)
  wall1, = ax.plot([-2, 2], [0, 0], 'b', lw=1)
  ball, = ax.plot([0], [0], 'ro', markersize=12)

  txt_title = ax.set_title('')

  ax.set_xlim(( -1.5, 1.5))
  ax.set_ylim((-0.5, 2.5))
  txt_title = ax.set_title('')
  interval = len(p_history)//num_frames
  def drawFrame(k):
    k = interval*k
    p0 = p_history[k]

    ball.set_data([p0[0]], [p0[1]])
    txt_title.set_text('t = {:.2f} sec'.format(t_history[k]))
    return ball,
  anim = animation.FuncAnimation(fig, drawFrame, frames=num_frames, interval=interval, blit=True)
  return anim
```

```python
# Animation code (No need to change)
anim = plot_bouncingball(p_hist, t_hist, num_frames= 100)
plt.close()
HTML(anim.to_html5_video())

# writervideo = animation.FFMpegWriter(fps=60)
# anim.save('ballbounding.mp4', writer=writervideo)
```

## Q.2 (a) [20 pts] Simulate the pin ball motion. Assume the radius of the ball is approximately zero and no frictional loss. Use coefficient of restitution, $\gamma$ = 0.9, initial pos: [0.5, 1.0], and initial velocity: [1.5, 0.5]. Simulate over 5 sec.

```python
# Collision checking code: No need to change
def point_line_distance(point, line):
  distance = np.linalg.norm(np.cross(line[0]-point, line[1]-point))/np.linalg.norm(line[1]-line[0])
  return distance

def detect_collision(pos, walls, threshold=0.001):
  dlist = []
  collision_idx = None
  for i, wall in enumerate(walls):
    d = point_line_distance(pos, wall)
    dlist.append(d)
    if d<=threshold:
      collision_idx = i
  return collision_idx
```

```python
# Simulate Pinball
def simulate_pinball(coeff_rest, init_pos, init_vel, delta_t, walls, wall_norms, wall_tan, N=6000):
  p_hist = np.zeros((N, 2))
  v_hist = np.zeros((N, 2))
  t_hist = np.zeros(N)

  p_hist[0, :] = init_pos
  v_hist[0, :] = init_vel
  t_hist[0] = 0
  for i in range(1, N):
    v_hist[i, :] = v_hist[i-1, :]

    # Put your code: Check collision and update velocity
    p_hist[i, :] = p_hist[i-1, :] + delta_t * v_hist[i-1, :]
    collision_idx = detect_collision(p_hist[i, :], walls)

    ## Handle Collision
    if collision_idx is not None:
      normal = np.array(wall_norms[collision_idx])
      tangent = np.array(wall_tan[collision_idx])

      v_nor = np.dot(v_hist[i, :], normal)
      v_tan = np.dot(v_hist[i, :], tangent)
      v_hist[i, :] = v_tan * tangent + coeff_rest * ( - v_nor * normal)
      p_hist[i, :] = p_hist[i-1, :] + delta_t * v_hist[i, :]

    t_hist[i] = i*delta_t
  return p_hist, v_hist, t_hist
```

```python
# Put your code: Simulate pinball
[p_hist_pinball, v_hist_pinball, t_hist_pinball] = simulate_pinball(coeff_rest=0.9, init_pos=np.array([0.5, 1.0]), init_vel=np.array([1.5, 0.5]), delta_t=0.001, walls=walls, wall_norms=wall_norms, wall_tan=wall_tan)
```

```python
# Visualize Pinball Matplotlib: No need to change
def plot_pinball(q, ts, walls, num_frames= 100):
  fig= plt.figure(figsize=(10,10))
  ax = plt.subplot(1,1,1)
  # m, c, l, g, mu, dt = params
  wall1, = ax.plot([walls[0][0][0], walls[0][1][0]], [walls[0][0][1], walls[0][1][1]], 'b', lw=1)
  wall2, = ax.plot([walls[1][0][0], walls[1][1][0]], [walls[1][0][1], walls[1][1][1]], 'b', lw=1)
  wall3, = ax.plot([walls[2][0][0], walls[2][1][0]], [walls[2][0][1], walls[2][1][1]], 'b', lw=1)
  wall4, = ax.plot([walls[3][0][0], walls[3][1][0]], [walls[3][0][1], walls[3][1][1]], 'b', lw=1)
  ball, = ax.plot([0], [0], 'ro', markersize=12)

  txt_title = ax.set_title('')

  ax.set_xlim(( -1, 2))
  ax.set_ylim((-0.5, 2.5))
  txt_title = ax.set_title('')
  interval = len(q)//num_frames
  def drawFrame(k):
    k = interval*k
    q0 = q[k]
    ball.set_data([q0[0]], [q0[1]])
    txt_title.set_text('t = {:.2f} sec'.format(ts[k]))
    return ball,
  anim = animation.FuncAnimation(fig, drawFrame, frames=num_frames, interval=interval, blit=True)
  return anim
```

## Q.3.(a) [20 pts] Implement a swing stick simulation.
Use parameters, [$m, I, c, l, \mu, dt$] = [1, 0.05, 0.5, 1.0, 0, 0.001].

Simulate the case when [$\theta_0, \dot{\theta}_0$] = [0, 0] and $F$ is applied at the tip of the arm during 0.05 sec to the vertical direction.


```python
def sim_step(theta, theta_dot, F, tau , params=None):
  m, I, c, l, mu, dt = params
  # Put your code: implement dynamics to compute the angular acceleration
  theta_ddot = (tau + F*l*np.cos(theta) - mu*theta_dot) / (I + m * c**2)

  # Fill your code: Semi-implciti Euler integration
  th_dot = theta_dot + theta_ddot * dt
  th = theta + th_dot * dt

  return [th, th_dot]

def simulate_stick(theta=0, theta_dot=0, F=50, tau=0, F_duration=0.05, F_start_time=0.0, T=1.0, params=None):
  dt = params[-1]
  ts = np.linspace(0, T, int(T/dt))
  theta_hist = [theta]
  theta_dot_hist = [theta_dot]

  # Put your code: Implement the case that the external force pushes the tip of the arm
  for i, t in enumerate(ts):
    if t <= F_duration:
      new_th, new_th_dot = sim_step(theta, theta_dot, 0, tau, params)
    else:
      new_th, new_th_dot = sim_step(theta, theta_dot, F, tau, params)
    # if t <= F_duration:
    #   new_th, new_th_dot = sim_step(theta_hist[i], theta_dot_hist[i], 0, tau, params)
    # else:
    #   new_th, new_th_dot = sim_step(theta_hist[i], theta_dot_hist[i], F, tau, params)
    # theta += new_th
    # theta_dot += new_th_dot
    # print(theta)
    # print(theta_dot)

    theta_hist.append(new_th)
    theta_dot_hist.append(new_th_dot)

  ts = ts.tolist()
  ts.append(T)
  return [theta_hist, theta_dot_hist, ts]
```

```python
# Visualization code: No need to change
def plot_stick(q, ts, params, num_frames= 100):
  fig= plt.figure(figsize=(10,10))
  ax = plt.subplot(1,1,1)
  m, I, c, l, mu, dt = params
  link1, = ax.plot([], [], 'b', lw=10)     # ax.plot returns a list of 2D line objects
  txt_title = ax.set_title('')

  ax.set_xlim(( -2.5, 2.5))    # Canvas size
  ax.set_ylim((-2.5, 2.5))
  txt_title = ax.set_title('')
  interval = len(q)//num_frames

  def drawFrame(k):
    k = interval*k
    q0 = q[k]
    rA = [l*np.cos(q0), l*np.sin(q0)]
    x1 = 0
    x2 = rA[0]
    y1 = 0
    y2 = rA[1]
    link1.set_data([x1, x2], [y1, y2])
    txt_title.set_text('t = {:.2f} sec'.format(ts[k]))
    return link1,

  anim = animation.FuncAnimation(fig, drawFrame, frames=num_frames, interval=interval, blit=True)
  return anim
```

```python
# Let's use param: [mass, Inertia, CoM, length, friction coefficient, delta_t]
# Put your code: Run simulation
params = [1, 0.05, 0.5, 1.0, 0, 0.001]
[theta_hist, theta_dot_hist, ts] = simulate_stick(params=params)
```
